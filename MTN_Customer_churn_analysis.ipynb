```python
# To sort according to total revenue
df.sort_values(by=['Total Revenue'],inplace=False,ascending=False)
```

```python
# Count of churn status
plt.figure(figsize=(5, 5))
status_counts = df['Customer Churn Status'].value_counts()

ax = sns.barplot(x=status_counts.index, y=status_counts.values, )
ax.set_xlabel('Churn Status')
ax.set_ylabel('Count')
ax.set_title('Customer Churn Status Count')

plt.tight_layout()
plt.show()
```

```python
# Calculate average satisfaction rate for churned customers for each churn reason

# Filter churned customers
churned_df = df[df['Customer Churn Status'].str.strip().str.lower() == 'yes']

# Calculate average satisfaction rate by reason
avg_satisfaction = churned_df.groupby('Reasons for Churn')['Satisfaction Rate'].mean().sort_values(ascending=False)

# Plot the average satisfaction rate by churn reason
plt.figure(figsize=(8, 4))
ax = sns.barplot(x=avg_satisfaction.values, y=avg_satisfaction.index, )
ax.set_xlabel('Average Satisfaction Rate')
ax.set_ylabel('Reason for Churn')
ax.set_title('Average Satisfaction Rate by Churn Reason')
plt.tight_layout()
plt.show()
```

```python
# Filter churned customers
churned_df = df[df['Customer Churn Status'].str.strip().str.lower() == 'yes']

# Count reasons for churn
reason_counts = churned_df['Reasons for Churn'].value_counts()
# Plot the top reasons for churn
plt.figure(figsize=((8, 4)))
sns.barplot(x=reason_counts.values, y=reason_counts.index,)
plt.xlabel('Count')
plt.ylabel('Reason for Churn')
plt.title('Top Reasons for Customer Churn')
plt.tight_layout()
plt.show()
```

```python
#customer churn rate and counts by age group
# Create age groups
df['Age_Group'] = pd.cut(df['Age'], 
                        bins=[0, 20, 30, 40, 50, 60, 100],
                        labels=['<20', '20-30', '31-40', '41-50', '51-60', '60+'])

# Calculate churn rate by age group
age_churn = pd.crosstab(df['Age_Group'], df['Customer Churn Status'], normalize='index') * 100
age_counts = pd.crosstab(df['Age_Group'], df['Customer Churn Status'])

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Churn rate plot
age_churn['Yes'].plot(kind='bar', ax=ax1, color='coral')
ax1.set_title('Churn Rate by Age Group')
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Churn Rate (%)')
ax1.tick_params(axis='x', rotation=45)

# Count plot
sns.countplot(data=df, x='Age_Group', hue='Customer Churn Status', ax=ax2, palette='viridis')
ax2.set_title('Customer Count by Age Group and Churn Status')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Count')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#The top 10 churn rate by state
# Create churn flag if not exists
if 'Churn_Flag' not in df.columns:
    df['Churn_Flag'] = df['Customer Churn Status'].str.strip().str.lower().apply(lambda x: 1 if x == 'yes' else 0)

# Group by state to calculate churn count and total customers
region_churn = df.groupby('State')['Churn_Flag'].sum().reset_index()
region_total = df.groupby('State')['Churn_Flag'].count().reset_index().rename(columns={'Churn_Flag':'Total_Customers'})
region_data = pd.merge(region_churn, region_total, on='State')
region_data['Churn_Rate'] = region_data['Churn_Flag'] / region_data['Total_Customers'] * 100

# Sort by the churn rate in descending order and select the top 10 states
top_10_states = region_data.sort_values('Churn_Rate', ascending=False).head(10)

print('Top 10 States by Churn Rate:')
print(top_10_states[['State', 'Churn_Rate']].to_string(index=False))

# Plot top 10 churn rate by state
plt.figure(figsize=(8, 4))
ax = sns.barplot(x='State', y='Churn_Rate', data=top_10_states, palette='magma')
ax.set_xlabel('State')
ax.set_ylabel('Churn Rate (%)')
ax.set_title('Top 10 States by Churn Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# create a heatmap to visualize the correlation between churn rate and age group

# First, we need to create a correlation matrix
correlation_matrix = df[['Age', 'Churn_Flag']].corr()

# Now, let's plot the heatmap
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap between Age and Churn Flag')
plt.show()
```
